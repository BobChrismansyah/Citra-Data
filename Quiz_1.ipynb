Hilangkan noise periodik pada gambar yang diberikan

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from scipy import fftpack
```

```python
im = plt.imread('lena_priodic.bmp').astype(float)

plt.figure()
plt.imshow(im, plt.cm.gray)
plt.title('Original image')
```

```python
#logaritma colormap
im_fft = fftpack.fft2(im)

def plot_spectrum(im_fft):
    plt.imshow(np.abs(im_fft), norm=LogNorm(vmin=5))
    plt.colorbar()
plt.figure()
plot_spectrum(im_fft)
plt.title('Transformasi Fourier')
```

```python
# Menyaring spectrun yang akan dihilangkan
fraction = 0.1

im_fft2 = im_fft.copy()

r, c = im_fft2.shape
im_fft2[int(r*fraction):int(r*(1-fraction))] = 0
im_fft2[:, int(c*fraction):int(c*(1-fraction))] = 0

plt.figure()
plot_spectrum(im_fft2)
plt.title('Filtered Spectrum')
```

```python
# Merekonstruksi gambar priodic dari spektrum yang disaring
im_new = fftpack.ifft2(im_fft2).real

plt.figure()
plt.imshow(im_new, plt.cm.gray)
plt.title('Hasil tanpa noise')
```
